```python
# This is the double gamma created by R's fmri library
# Found it on a video created by Mumford on parametric modulation
# Link: https://www.youtube.com/watch?v=2W7Rso-4Hqg
```

```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
a1 = 6; a2 = 12; b1 = 0.9; b2 = 0.9
```

```python
d1 = a1 * b1; d2 = a2 * b2
```

```python
x = np.array(range(100))
```

```python
c1 = (x / d1) ** a1
```

```python
plt.plot(c1)
```

```python
c2 = 0.35 * (x / d2) ** a2
```

```python
plt.plot(c2)
```

```python
m1 = np.exp(-(x - d1) / b1)
```

```python
plt.plot(m1)
```

```python
m2 = np.exp(-(x - d2) / b2)
```

```python
plt.plot(m2)
```

```python
plt.plot(c1 * m1, 'r')
plt.plot(c2 * m2, 'b')
plt.plot(c1*m1 - c2*m2, 'g')
```

```python
res = c1 * np.exp(-(x - d1) / b1) - c2 * np.exp(-(x - d2) / b2)
```

```python
plt.plot(res)
```
